```python
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use("dark_background")
%matplotlib inline
```

```python
iq = np.load("image_quality.pkl",allow_pickle=True)
iq[0]
```

```python
n_mode = [x["n_mode"] for x in iq]
```

```python
fig, ax2 = plt.subplots(figsize=(10,7)) 
h21 = ax2.plot(n_mode,[x["pssn"]["v_band"] for x in iq],"C0o:",label="V")
h22 = ax2.plot(n_mode,[x["pssn"]["h_band"] for x in iq],"C1o:",label="H")
ax2.hlines(0.9969,14,42,"C0")
ax2.hlines(0.9981,14,42,"C1")
ax2.grid()
ax2.set_xlabel("Bending mode #")
ax2.set_ylabel("PSSn")
ax1= ax2.twinx()
h11 = ax1.plot(n_mode,[x["wfe_rms"] for x in iq],'ko:',alpha=0.25,label="WFE RMS")
ax1.set_ylabel("WFE RMS[nm]")
ax1.legend(loc=7)
ax2.legend(loc="center")
```
